   <font color="brown"> Hint: You can look into this blog for more information. https://towardsdatascience.com/named-entity-recognition-with-nltk-and-spacy-8c4a7d88e7da <font>

```python
from nltk import ne_chunk 

tagged = nltk.pos_tag(tokens)
namedEnt = nltk.ne_chunk(tagged)
table= PrettyTable(["Entity", "Label"])

for e in namedEnt:
    try:
        table.add_row(["".join(c[0] for c in e), e.label()])
    except AttributeError:
        pass

table
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("""As President Donald Trump orders federal workers to return to the office, he spent the fourth straight day at home here at his Mar-a-Lago resort.""")

table = PrettyTable()
table.field_names = ["Word", "POS", "is_stop?", "Entity"]
for token in doc:
    table.add_row([token.text, token.pos_, token.is_stop, token.ent_type_])
table
```
